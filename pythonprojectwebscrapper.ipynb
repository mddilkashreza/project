```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url)
soup.text
raw_code =  BeautifulSoup(soup.text, "lxml")
raw_code
```

```python
table_code = raw_code.table
```

```python
tr_tags = table_code.find_all('tr')
tr_tags
```

```python
l = []
for tag in tr_tags:
    x = tag.text.split('\n')
    if x[1] != "":
        l.append(x[1:])
    
    
print(l)
```

```python
import csv
file = open('covid_data.csv', 'w')
y = csv.writer(file)
y.writerows(l)
file.close()
```
